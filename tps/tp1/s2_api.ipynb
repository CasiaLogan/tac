## Imports

```python
import json
import requests
import os
```

## Utiliser [Wikidata](https://publicapis.io/wikidata-api) pour connaître le context historique des textes

```python
def fetch_wikidata(Q1190356, cache_dir="dataTP/wikicache"): #utilise le wikidata QID pour récupérer les données sur l'RTBF et definir un répertoire de cache
    cache_file = os.path.join(cache_dir, f"{Q1190356}.json") 
    if os.path.exists(cache_file):
        with open(cache_file, "r") as f:
            data = json.load(f)
            print(json.dumps(data, indent=2))
            return data

    url = "https://www.wikidata.org/w/api.php" #get the data from wikidata API
    params = {
        "action": "wbgetentities",
        "ids": Q1190356,
        "format": "json",
        "props": "labels|descriptions"
    }
    headers = {"User-Agent": "TestForSTICB545/3.0 (casia.logan@ulb.be)"} #doit etre unique pour que wikidata n'ignore pas la requete (regles de wikidata api)

    response = requests.get(url, params=params, headers=headers)
    data = response.json()

    os.makedirs(cache_dir, exist_ok=True)
    with open(cache_file, "w") as f:
        json.dump(data, f)

    print(json.dumps(data, indent=2))
    return data

data = fetch_wikidata("Q1190356")
```
